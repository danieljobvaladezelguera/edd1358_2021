```
#readline()
archivo = open('/content/sample_data/frutas', 'rt')

data = archivo.readline()
while data != "":
  print(data)
print(archivo.readline())

archivo.close
```

```
# readlines()

archivo = open ('/content/sample_data/frutas', 'rt')

data = archivo.readlines()
print(data)

frutas =[]
for linea in data:
    frutas.append(linea.strip())

frutas_final = []
for elem in frutas:
    for fruta in elem.split(','):
        frutas_final.append(fruta.strip())

print(frutas)

archivo.close()
```
